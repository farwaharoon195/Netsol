````python
import numpy as np
import matplotlib.pyplot as plt

subjects = ['Math', 'Science', 'English', 'History']
class_a  = [72, 85, 78, 65]
class_b  = [68, 80, 74, 70]

# Number of subjects
n = len(subjects)

# Create x positions
x = np.arange(n)

# Width of the bars
width = 0.35

# Plot Class A bars (offset left)
plt.bar(x - width/2, class_a, width, label='Class A', color='skyblue')

# Plot Class B bars (offset right)
plt.bar(x + width/2, class_b, width, label='Class B', color='salmon')

# Add value labels above each bar
for i in range(n):
    plt.text(x[i] - width/2, class_a[i] + 1, str(class_a[i]), ha='center')
    plt.text(x[i] + width/2, class_b[i] + 1, str(class_b[i]), ha='center')

# Add title, axis labels, legend, and x-tick labels
plt.title('Class Scores by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.xticks(x, subjects)
plt.legend()

# Show plot
plt.show()
````

````python
import matplotlib.pyplot as plt

months     = ['Jan','Feb','Mar','Apr','May','Jun']
sales      = [200, 240, 180, 300, 350, 320]

cities     = ['Karachi','Lahore','Islamabad','Peshawar']
population = [16, 13, 2, 4]   # millions

hours      = [1, 2, 3, 4, 5, 6, 7, 8]
grades     = [50, 55, 60, 65, 72, 78, 85, 90]

categories = ['Food', 'Rent', 'Transport', 'Entertainment']
expenses   = [40, 30, 15, 15]

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Top-left: line plot of sales over months
axes[0, 0].plot(months, sales, marker='o', color='skyblue', linewidth=2)
axes[0, 0].set_title('Sales Over Months')
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Sales')

# Top-right: bar chart of city populations
axes[0, 1].bar(cities, population, color='salmon', edgecolor='black')
axes[0, 1].set_title('City Populations (millions)')
axes[0, 1].set_ylabel('Population')

# Bottom-left: scatter plot of hours vs grades
axes[1, 0].scatter(hours, grades, color='green', s=100)
axes[1, 0].set_title('Hours vs Grades')
axes[1, 0].set_xlabel('Hours Studied')
axes[1, 0].set_ylabel('Grades')

# Bottom-right: pie chart of expenses
axes[1, 1].pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90, colors=['gold','lightblue','lightgreen','lightcoral'])
axes[1, 1].set_title('Monthly Expenses')

fig.suptitle("Monthly Dashboard", fontsize=16)
plt.tight_layout()
plt.show()
````

````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool, ColumnDataSource

output_notebook()

days     = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 180, 150, 200, 250, 300, 270]

# YOUR CODE HERE
# 1. Create a ColumnDataSource


# 2. Create a Bokeh figure with title and axis labels


# 3. Add a line renderer


# 4. Add circle markers


# 5. Add a HoverTool


# 6. Show the plot
````

````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool, ColumnDataSource

output_notebook()

days     = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 180, 150, 200, 250, 300, 270]

# 1. Create a ColumnDataSource
source = ColumnDataSource(data=dict(day=days, visitors=visitors))

# 2. Create a Bokeh figure
p = figure(title="Website Visitors Over a Week",
           x_range=days,  # categorical x-axis
           y_axis_label="Number of Visitors",
           x_axis_label="Day",
           plot_width=700, plot_height=400,
           tools="pan,wheel_zoom,box_zoom,reset")

# 3. Add a line renderer
p.line('day', 'visitors', source=source, line_width=3, color='navy', legend_label="Visitors")

# 4. Add circle markers
p.circle('day', 'visitors', source=source, size=10, color='orange', legend_label="Visitors")

# 5. Add a HoverTool
hover = HoverTool(tooltips=[("Day", "@day"), ("Visitors", "@visitors")])
p.add_tools(hover)

p.legend.location = "top_left"
p.legend.click_policy="hide"

# 6. Show the plot
show(p)
````
