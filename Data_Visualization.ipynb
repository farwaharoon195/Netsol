# Introduction to Data Visualization in Python


## Why is Data Visualization Important?

Data visualization is a key skill in data analysis, data science, and communication. Here's why it matters:

### 🧠 1. Helps You Understand Data Better
Raw data in spreadsheets or tables can be hard to interpret. Visualization transforms rows and columns of numbers into **visual meaning**, making it easier to:

- Spot **trends** and **patterns**
- Identify **outliers** and **anomalies**
- Understand the **distribution** of values

> Example: A sales report across months might just look like numbers, but a line chart can instantly show upward or downward trends.

---

### 📢 2. Communicates Insights Clearly

Visuals make it easier to share findings with both **technical and non-technical stakeholders**. Instead of describing relationships in words, a good graph shows the story in seconds.

> Think of dashboards in business — they condense complex performance metrics into charts that anyone can understand at a glance.

---

### ⌛ 3. Saves Time in Decision Making

Data visualization **reduces time spent on analysis** and **speeds up decisions** by:

- Highlighting key issues immediately
- Allowing interactive filtering and zooming (in tools like Plotly or Bokeh)
- Making it easier to compare multiple variables or categories simultaneously

---

### ❌ What Happens If We Don't Visualize Data?

Without visualization:

- You may **miss important insights** buried in raw data
- Errors and outliers may go **undetected**
- Stakeholders may **misinterpret** findings due to poor communication
- You spend **more time exploring** instead of concluding

---

### 🔍 Bottom Line

> **"A picture is worth a thousand rows."**

Data visualization turns complex datasets into visual narratives, improving **comprehension**, **communication**, and **decision-making speed**.

## 📈 Line Plot

### ✅ When to Use:
- To show **trends over time**
- To track changes or progression
- Often used in time series analysis

> Example: Stock prices, temperature change, sales over months



````python
# 🧪 Example: Temperature Over a Week

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temps = [21, 23, 19, 24, 22, 27, 26]

plt.plot(days, temps, marker='^', linestyle='-', color='teal')
plt.title('Temperature Over a Week')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
````

````python
import matplotlib.pyplot as plt
# Change line to solid and color to red
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temps = [21, 23, 19, 24, 22, 27, 26]

plt.plot(days, temps, marker='o', linestyle='-', color='red')
plt.title('Test 1: Solid Red Line with Square Markers')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()




# # Use triangle markers and remove grid
plt.plot(days, temps, marker='^', linestyle='--', color='green')
plt.title('Test 2: Green Dashed Line with Triangles')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.grid(False)
plt.show()




# # Add two temperature lines (e.g., this week vs last week)
last_week = [20, 22, 18, 23, 21, 25, 24]

plt.plot(days, temps, marker='o', linestyle='--', label='This Week', color='blue')
plt.plot(days, last_week, marker='s', linestyle='-', label='Last Week', color='orange')
plt.title('Test 3: Temperature Comparison')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()



# Reverse the temperature list
plt.plot(days, temps[::-1], marker='D', linestyle='-', color='purple')
plt.title('Test 4: Reversed Temperature Data')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()



# # Custom figure size and font sizes
plt.figure(figsize=(10, 5))
plt.plot(days, temps, marker='o', linestyle='--', color='teal')
plt.title('Test 5: Custom Figure Size', fontsize=16)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.grid(True)
plt.show()
````

## 📊 Bar Chart

### ✅ When to Use:
- To **compare categories**
- Good for **frequency counts**, survey responses, or group comparisons

> Example: Number of students per class, sales per region, product ratings


````python
# 🧪 Example: Clinic Visits Per Day
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visits = [20, 30, 30, 25, 10, 5, 15]

plt.bar(days, visits, color='orange')
plt.title('Clinic Visits Per Day')
plt.xlabel('Day')
plt.ylabel('Number of Visits')
plt.show()
````

````python
# # Change bar color to blue and adjust bar width
plt.bar(days, visits, color='blue', width=0.5)
plt.title('Test 1: Blue Bars with Narrow Width')
plt.xlabel('Day')
plt.ylabel('Number of Visits')
plt.show()




# Flip the bars horizontally
plt.barh(days, visits, color='green')
plt.title('Test 2: Horizontal Bar Chart')
plt.xlabel('Number of Visits')
plt.ylabel('Day')
plt.show()



# Add numeric labels above each bar
bars = plt.bar(days, visits, color='orange')
plt.title('Test 3: Clinic Visits with Labels')
plt.xlabel('Day')
plt.ylabel('Number of Visits')

# Add value labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval , yval, ha='center', va='bottom')

plt.show()




# # Widen the figure and rotate x-axis labels
plt.figure(figsize=(10, 5))
plt.bar(days, visits, color='purple')
plt.title('Test 4: Rotated Labels with Custom Size')
plt.xlabel('Day')
plt.ylabel('Number of Visits')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()




# Compare current week vs previous week
import numpy as np

previous_week = [15, 30, 28, 20, 12, 10, 18]
x = np.arange(len(days))

plt.bar(x - 0.2, visits, width=0.4, label='This Week', color='teal')
plt.bar(x + 0.2, previous_week, width=0.4, label='Last Week', color='coral')
plt.xticks(x, days)
plt.title('Test 5: Weekly Comparison of Visits')
plt.xlabel('Day')
plt.ylabel('Number of Visits')
plt.legend()
plt.show()
````

````python
# # Use different marker shape and color
study_hours = [1, 2, 3, 4, 5, 6, 7]
exam_scores = [50, 55, 60, 63, 70, 75, 80]
plt.scatter(study_hours, exam_scores, color='red', marker='^')
plt.title('Test 1: Triangle Markers (Red)')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.grid(True)
plt.show()




# Bubble size varies with hours studied (just for demo)
sizes = [s * 20 for s in study_hours]

plt.scatter(study_hours, exam_scores, color='teal', s=sizes)
plt.title('Test 2: Bubble Size Based on Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.grid(True)
plt.show()






plt.scatter(study_hours, exam_scores, color='orange')

 # Add labels for each point
for i in range(len(study_hours)):
     plt.text(study_hours[i] + 0.1, exam_scores[i], f"{exam_scores[i]}")

plt.title('Test 3: Annotated Points')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.grid(True)
plt.show()



colors = ['green' if score >= 60 else 'red' for score in exam_scores]

plt.scatter(study_hours, exam_scores, color=colors)
plt.title('Test 5: Color by Pass/Fail (Green ≥ 60)')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.grid(True)
plt.show()

````

## 📊 Histogram

### ✅ When to Use:
- To show the **distribution** of a single numeric variable
- Helps identify **frequency**, **skewness**, **normality**, or **outliers**

> Example: Age distribution, test score frequency, income distribution


````python
# 🧪 Example: Age Distribution of Survey Respondents
ages = [22, 25, 26, 22, 30, 28, 24, 32, 35, 28, 24, 27, 40, 42, 30]

plt.hist(ages, bins=2, color='purple', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.xticks(range(min(ages), max(ages)+1, 2))
plt.ylabel('Number of People')
plt.show()
````

````python
# Test 2: More bins (finer granularity)
plt.hist(ages, bins=10, color='teal', edgecolor='black')
plt.title('More Bins (10)')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Test 3: Horizontal histogram

plt.hist(ages, bins=5, color='orange', edgecolor='black', orientation='horizontal')
plt.title('Horizontal')
plt.xlabel('Count')
plt.ylabel('Age')

plt.tight_layout()
plt.show()

#on basis of range we define bins
````

## 📦 Box Plot

### ✅ When to Use:
- To visualize **summary statistics** like median, quartiles, and outliers
- Great for comparing **distributions** across multiple groups

> Example: Compare salaries of different departments, marks of students across sections


````python
# 🧪 Example: Test Scores of Two Classes
class_A = [55, 60, 65, 70, 75, 80, 85]
class_B = [58, 62, 67, 73, 78, 85, 88]

plt.boxplot([class_A, class_B], tick_labels=['Class A', 'Class B'])
plt.title('Score Distribution by Class')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
````

````python
# 🔹 Test 2: Show Mean on Boxplot
class_A = [55, 60, 65, 70, 75, 80, 85]
class_B = [58, 62, 67, 73, 78, 85, 88]
plt.boxplot([class_A, class_B], labels=['Class A', 'Class B'], showmeans=True)
plt.title('Test 2: Boxplot with Mean')
plt.ylabel('Marks')
plt.grid(True)
plt.show()

# 🔹 Test 3: Horizontal Boxplot
plt.boxplot([class_A, class_B], labels=['Class A', 'Class B'], vert=False)
plt.title('Test 3: Horizontal Boxplot')
plt.xlabel('Marks')
plt.grid(True)
plt.show()

# 🔹 Test 4: Styled Boxplot with Colors
plt.boxplot([class_A, class_B],
           labels=['Class A', 'Class B'],
           patch_artist=True,
           boxprops=dict(facecolor='lightblue'),
           medianprops=dict(color='red', linewidth=2),
           showmeans=True,
           meanprops=dict(marker='o', markerfacecolor='black', markersize=6))

plt.title('Test 4: Styled Boxplot with Mean')
plt.ylabel('Marks')
plt.grid(True)
plt.show()

# 🔹 Test 5: Unequal Data Lengths
class_C = [60, 65, 70]
plt.boxplot([class_A, class_B, class_C], labels=['Class A', 'Class B', 'Class C'])
plt.title('Test 5: Boxplot with Unequal Lengths')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
````

## 📐 What are Subplots and Layouts?

### ✅ What is a Subplot?

A **subplot** is a way to show multiple plots (charts) in a single figure window.

- Think of it as creating a **grid of plots** — like a collage of charts.
- This helps compare multiple variables or visualizations side-by-side.

### 🧱 Layout

The layout refers to how the subplots are **arranged in rows and columns**.

### 🔧 Common Function:
```python
plt.subplot(n_rows, n_columns, index)

- n_rows: total number of rows
- n_columns: total number of columns
- index: plot position (starting from 1)

Another more flexible way (recommended for modern code) is:


fig, axes = plt.subplots(n_rows, n_columns)

````python


### 🧪 Cell: Example – Using `plt.subplot()`

# Example: Displaying a line plot and bar chart side by side
import matplotlib.pyplot as plt

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
temps = [21, 23, 19, 24, 22]
visits = [20, 35, 30, 25, 10]

plt.figure(figsize=(10, 4))

# Line Plot
plt.subplot(1, 2, 2)
plt.plot(days, temps, marker='o')
plt.title('Temperature')

# Bar Chart
plt.subplot(1, 2, 1)
plt.bar(days, visits, color='orange')
plt.title('Clinic Visits')

plt.tight_layout()
plt.show()
````

````python
# Example: Using subplots() for better control
fig, axes = plt.subplots(2, 2, figsize=(10, 8))





# Line Plot
axes[0, 0].plot([1, 2, 3], [10, 20, 30])
axes[0, 0].set_title('Line Plot')

# Bar Chart
axes[0, 1].bar(['A', 'B', 'C'], [5, 7, 3])
axes[0, 1].set_title('Bar Chart')

# Histogram
axes[1, 0].hist([22, 25, 25, 30, 28, 28, 35], bins=3)
axes[1, 0].set_title('Histogram')

axes[1, 1].scatter([1, 2, 3], [4, 5, 6])
axes[1, 1].set_title('Scatter Plot')

plt.tight_layout()
plt.show()
````

## 📊 Seaborn Bar Plot

### ✅ When to Use:
- To compare **means** or **aggregated values** across categories
- Automatically computes **confidence intervals**

> Example: Average bill amount by gender


````python
sns.barplot(x='sex', y='total_bill', data=tips)
plt.title('Average Total Bill by Gender')
plt.show()
````

````python
# 🔹 Test 2: Add Hue for Time (Lunch/Dinner comparison)
sns.barplot(x='sex', y='total_bill', hue='time', data=tips)
plt.title('Test 2: Average Total Bill by Gender & Time')
plt.show()

# 🔹 Test 3: Horizontal Bar Plot
sns.barplot(y='sex', x='total_bill', data=tips)
plt.title('Test 3: Horizontal Bar Plot')
plt.show()

# 🔹 Test 4: Use Estimator = sum instead of mean
import numpy as np
sns.barplot(x='day', y='total_bill', data=tips, estimator=np.sum)
plt.title('Test 4: Total Bill by Day (Sum)')
plt.show()

# # 🔹 Test 5: Customize Color Palette
sns.barplot(x='day', y='tip', data=tips, hue='day', palette='Set1', legend=True)
plt.title('Test 5: Tip Amount by Day (Custom Colors)')
plt.show()
````

## 📈 Seaborn Line Plot

### ✅ When to Use:
- To show **trends over time** or a continuous variable
- Automatically computes **confidence interval** shading

> Example: Trend of tips by size of group


````python
sns.lineplot(x='size', y='tip', data=tips)
plt.title('Tip vs Group Size')
plt.show()
````

````python
# 🔹 Test 2: Add Hue (line per time: Lunch/Dinner)
sns.lineplot(x='size', y='tip', data=tips, hue='time')
plt.title('Test 2: Tip vs Group Size by Time')
plt.show()

# 🔹 Test 3: Use Median instead of Mean
sns.lineplot(x='size', y='tip', data=tips, estimator=np.median)
plt.title('Test 3: Median Tip vs Group Size')
plt.show()

# 🔹 Test 4: Show Confidence Interval (default: 95%)
sns.lineplot(x='size', y='tip', data=tips, ci=95)
plt.title('Test 4: Tip vs Group Size with CI=95%')
plt.show()

````

## 🔴 Seaborn Scatter Plot

### ✅ When to Use:

- Use `hue` to add a categorical third dimension

````python
# 🔹 Test 2: Add Style to Differentiate Points
sns.scatterplot(x='total_bill', y='tip', hue='sex', style="time", data=tips)
plt.title('Test 2: Tip vs Total Bill (Color by Gender, Shape by Time)')
plt.show()

# 🔹 Test 3: Change Size of Points by Group Size
sns.scatterplot(x='total_bill', y='tip', size='size', hue='sex', data=tips)
plt.title('Test 3: Tip vs Total Bill (Point Size = Group Size)')
plt.show()

# 🔹 Test 4: Customize Palette and Markers
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips, palette='Set1', marker='D')
plt.title('Test 4: Custom Colors and Markers')
plt.show()

# 🔹 Test 5: Add Regression Line with Scatterplot
sns.regplot(x='total_bill', y='tip', data=tips, scatter=True)
plt.title('Test 5: Tip vs Total Bill with Regression Line')
plt.show()
````

## 🔴 Bokeh Example: Interactive Scatter Plot

### ✅ When to Use:
- To explore relationships between variables with hover tooltips

> Example: Income vs Age


````python
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool

# IMPORTANT for notebook
output_notebook()

# Data
x = [20, 25, 30, 35, 40, 45]
y = [20000, 25000, 27000, 30000, 32000, 35000]

# Data source
source = ColumnDataSource(data=dict(age=x, income=y))

# Figure
p = figure(title="Age vs Income", x_axis_label='Age', y_axis_label='Income',
           tools="pan,wheel_zoom,box_zoom,reset")

# Scatter (UPDATED method)
p.scatter(x='age', y='income', size=10, color="navy", alpha=0.6, source=source)

# Hover tool (CORRECT way)
hover = HoverTool(tooltips=[
    ("Age", "@age"),
    ("Income", "@income")
])

p.add_tools(hover)

# Show plot
show(p)
````

## 📈 Plotly Line Plot

### ✅ When to Use:
- To show **trends over time** with zooming and hover interaction

> Example: Stock prices or weather changes

````python
import plotly.express as px

# Sample data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
temp = [22, 25, 24, 23, 26]

fig = px.line(x=days, y=temp, title='Temperature Over the Week',
              labels={'x': 'Day', 'y': 'Temperature (°C)'})
fig.show()
````

````python
# 🔹 Test 2: Add Markers to Line
fig = px.line(x=days, y=temp, title='Test 2: Line Plot with Markers',
              labels={'x': 'Day', 'y': 'Temperature (°C)'})
fig.update_traces(mode='lines+markers', marker=dict(size=10, color='red'))
fig.show()

# 🔹 Test 3: Custom Line Style
fig = px.line(x=days, y=temp, title='Test 3: Custom Line Style',
           labels={'x': 'Day', 'y': 'Temp (°C)'})
fig.update_traces(line=dict(dash='dashdot', width=3, color='green'))
fig.show()

# 🔹 Test 4: Show Hover Labels Always
fig = px.line(x=days, y=temp, title='Test 4: Always Show Hover Labels',
               labels={'x': 'Day', 'y': 'Temperature'})
fig.update_traces(mode='lines+markers', hoverinfo='x+y', hoverlabel=dict(bgcolor="white"))
fig.show()

````

## 📊 Plotly Bar Chart

## 📚 Summary: Comparing Visualization Libraries

Here's a quick comparison of the four major Python visualization libraries we've covered:

| Library     | Type        | Strengths                                                                 | Best Use Cases                             |
|-------------|-------------|---------------------------------------------------------------------------|--------------------------------------------|
| **Matplotlib** | Static       | Full control, low-level API, highly customizable                        | Publications, detailed plots, subplots     |
| **Seaborn**    | Static       | Built on Matplotlib, beautiful defaults, easy for statistical plots     | Quick EDA, distributions, correlations      |
| **Bokeh**      | Interactive  | Web-ready plots, interactivity (hover, zoom, filters), good layout tools| Dashboards, web apps, data exploration      |
| **Plotly**     | Interactive  | Very interactive, attractive UI, works with Dash, 3D and map support    | Dashboards, storytelling, 3D visualizations |

### 🔍 Key Takeaways

- Use **Matplotlib** when you need precision, customization, and multi-plot control.
- Use **Seaborn** for quick and beautiful statistical visualizations with fewer lines of code.
- Use **Bokeh** if you want browser-based **interactivity** and layout control.
- Use **Plotly** for modern, interactive dashboards or when working with 3D or map-based data.

> 🔧 Choose the right tool based on your audience (technical vs. non-technical), need for interactivity, and the complexity of your data.
